### Variables Description

The dataset contains multiple variables that describe different aspects of terrorist incidents.

- city: city in which the attack happened.
- attacktype1_txt: type of attack carried out, such as bombing or armed assault.
- weaptype1_txt: primary weapon type used in the attack.
- targtype1_txt: type of target involved in the incident.
- gname: name of the terrorist group responsible for the attack, if known.
- nkill: number of people killed in the incident.
- nwound: number of people wounded in the incident.

These variables help in analyzing the patterns, trends, and impact of terrorist activities across

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# making a copy of the original dataset
df_clean = df.copy()

# handling missing values for important numerical columns
df_clean['nkill'] = df_clean['nkill'].fillna(0)
df_clean['nwound'] = df_clean['nwound'].fillna(0)

# replacing missing text values with 'unknown'
df_clean['country_txt'] = df_clean['country_txt'].fillna('unknown')
df_clean['region_txt'] = df_clean['region_txt'].fillna('unknown')
df_clean['city'] = df_clean['city'].fillna('unknown')
df_clean['attacktype1_txt'] = df_clean['attacktype1_txt'].fillna('unknown')
df_clean['weaptype1_txt'] = df_clean['weaptype1_txt'].fillna('unknown')
df_clean['targtype1_txt'] = df_clean['targtype1_txt'].fillna('unknown')
df_clean['gname'] = df_clean['gname'].fillna('unknown')

# checking if missing values still exist
df_clean.isnull().sum()
````

### What all manipulations have you done and insights you found?

During the data wrangling process, I performed a few basic but important steps to prepare the
dataset for analysis. First, I created a copy of the original dataset so that the raw data
remains unchanged and can be referred to if needed.

I checked the dataset for duplicate records and found that there were no duplicate rows, which
indicates that each terrorist incident is uniquely recorded. I also examined the structure of
the dataset to understand the number of rows, columns, data types, and overall size of the data.

To handle missing values, I focused on the key variables that are important for analysis. For
numerical columns such as the number of people killed and wounded, I replaced missing values with
zero to avoid issues during calculations. For categorical variables like country, region, city,
attack type, weapon type, target type, and terrorist group name, I replaced missing values with
the label 'unknown' to maintain consistency in the data.

From the initial exploration, I observed that the dataset is large and contains a mix of numerical
and categorical variables. While some columns contain a high number of missing values, the main
variables required for exploratory analysis are sufficiently populated. These observations
confirm that the dataset is suitable for exploratory data analysis and can provide meaningful
insights into global terrorism patterns.

#### Chart - 10

````python
# Chart - 10 visualization code
# selecting top regions for clarity
top_regions = df_clean['region_txt'].value_counts().head(5).index

# filtering data for top regions
region_year_data = df_clean[df_clean['region_txt'].isin(top_regions)]

# creating pivot table
region_year_pivot = pd.pivot_table(
    region_year_data,
    index='iyear',
    columns='region_txt',
    aggfunc='size',
    fill_value=0
)

# plotting stacked bar chart
region_year_pivot.plot(
    kind='bar',
    stacked=True,
    figsize=(12, 6)
)

plt.xlabel('year')
plt.ylabel('number of attacks')
plt.title('region wise terrorist attacks over the years')
plt.legend(title='region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

I selected a stacked bar chart because it allows comparison of multiple categories over time.
This chart helps visualize how terrorist attacks are distributed across major regions for each
year, while also showing overall trends.

Using a stacked bar chart makes it easier to compare both individual regional contributions and

#### Chart - 11

````python
# Chart - 11 visualization code
# creating a cross table between region and attack type
region_attack_ct = pd.crosstab(
    df_clean['region_txt'],
    df_clean['attacktype1_txt']
)

# selecting top regions and attack types for clarity
top_regions = region_attack_ct.sum(axis=1).sort_values(ascending=False).head(6).index
top_attacks = region_attack_ct.sum(axis=0).sort_values(ascending=False).head(6).index

heatmap_data = region_attack_ct.loc[top_regions, top_attacks]

# plotting heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt='d')
plt.title('heatmap of attack types across regions')
plt.xlabel('attack type')
plt.ylabel('region')
plt.show()
````

##### 1. Why did you pick the specific chart?

I selected a heatmap because it is useful for analyzing relationships between two categorical
variables at the same time. In this case, the chart helps understand how different attack types
are distributed across various regions.

The heatmap format makes it easy to identify high and low concentration areas using color
intensity.
